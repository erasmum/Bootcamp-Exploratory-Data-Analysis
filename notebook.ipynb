# 4.	Which region has the highest % of population over 60?

# 5.	Does fertility decrease as income increases? Are there any countries that donâ€™t seem to follow this relation?

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv(r"c:\users\erasmum\code\Data2.csv")
selcol = df[['FertilityRate','GNI']]
dfcorr = selcol.copy()
corrMat = dfcorr.corr()
sns.heatmap(corrMat, annot=True)
plt.show()
```

As per the plot above, there is a correlation between increased income and lower fertility rates.  When looking at the data itself, this is not a linear trend.  A country such as Qatar, who has the highest income, does have a slightly higher fertility rate thatn some contries with lower income.

# 8.	What is the life expectancy across different regions? Create a box-and-wiskers plot to investigate this. What can we conclude about life expectancy across different regions? 

```python
import pandas as pd
import seaborn as sns
sns.set_theme()
df = pd.read_csv(r"c:\users\erasmum\code\Data2.csv")
ax = sns.boxplot(x = 'LifeExpectancy', y = 'Region', data=df)
```

We can see that the life expectancy varies significantly across regions.  
When we look at the Eastern Mediterranean, we can see that there is a large spread between the countries in the region.  
Similarly, we can see that Africa also has a large spread, but the median is much lower than the other regions and their minimum is the lowest of all regions.  There is also an outlier.  Their maximum does not even reach the median of Europe and the Americas.
Europe has the highest median and a smaller spread in the data.  They do also have an outlier.
The Americas have the second highest median and the smallest spread.  They do however have 2 outliers, of which one is significantly lower than the minimum.
